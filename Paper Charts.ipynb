```python
import pandas as pd
import matplotlib
import seaborn as sns

metrics = pd.DataFrame(
    [
     ["LSTM",       78.89, "Recall Elevated"],
     ["RuleMatrix", 43.22, "Recall Elevated"],
     ["LSTM",       41.31, "Precision Elevated"],
     ["RuleMatrix", 17.30, "Precision Elevated"],
     ["LSTM",       99.63, "Recall Normal"],
     ["RuleMatrix", 99.33, "Recall Normal"],
     ["LSTM",       99.93, "Precision Normal"],
     ["RuleMatrix", 99.81, "Precision Normal"],
    ],
    columns=["Model", "Metric", "Score"]
)

plt.figure()

sns.set_style("white")

matplotlib.use("pgf")
matplotlib.rcParams.update({
    "pgf.texsystem": "pdflatex",
    'font.family': 'serif',
    'text.usetex': True,
    'pgf.rcfonts': False,
})

sns.barplot(x="Score", y="Metric", hue="Model", data = metrics)
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))

plt.xlabel("")
plt.ylabel("")
plt.title("Performance Metrics for Predicting CSO Events")
plt.xticks(rotation=45)

plt.savefig('Dataset-Analysis/charts/performance_metrics.pdf', bbox_inches='tight')
```
